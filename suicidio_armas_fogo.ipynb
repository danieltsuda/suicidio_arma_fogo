# Análise de dados de suicídios por armas de fogo no Brasil

Dataset retirado do site kaggle:

https://www.kaggle.com/datasets/istanislau/suicdios-por-armas-de-fogo-no-brasil

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dados_ipea.csv', sep=';')
```

```python
df.head()
```

```python
# exclusão de coluna que não será utilizada nesta an
df = df.drop(columns=['cod'])
```

```python
# renomeando colunas
df = df.rename(columns={'nome': 'Estado', 'período': 'Ano', 'valor': 'Valor'})
```

```python
df.head()
```

```python
# criando lista com os anos do dataframe
lista_ano = list(df['Ano'].unique())
lista_ano
```

```python
# criando função para contagem dos suicídios por ano
def contagem(ano):
    x = df.loc[df['Ano'] == ano, 'Valor'].sum()
    print(x)
```

```python
# loop for aplicando a função criada percorrendo a lista de anos
for ano in lista_ano:
    contagem(ano)
```

É possível notar que, ao longo dos anos, a taxa de suicídio por armas caiu, provavelmente devido a implementação de normas mais 
rigorosas para se adquirir uma arma de fogo.

```python
# função para criar gráficos para comparar a taxa de suicídio de cada estado por ano
def auto_plot(ano):
    df_func = df.loc[df['Ano'] == ano]
    fig, ax = plt.subplots(figsize=(15,10))
    sns.barplot(data=df_func, x='Estado', y='Valor', ax=ax).set_title('Contagem por Estados no ano de ' + str(ano))
    plt.show()
    fig.savefig(str(ano))
```

```python
# loop for aplicando a função acima percorrendo a lista de ano
for ano in lista_ano:
    auto_plot(ano)
```

Os estados com a maior taxa de suicídio ao longo dos anos são: SP e RS, seguidos por MG e PR.
